```python
#Numpy is used for Numerical Computing and is foundational for understanding Tensors
import numpy as np

print("Numpy imported successfully")

# 1D Array
numbers = np.array([1,2,3,4])
print("1-Dimensional Array" ,numbers)
print(type(numbers))
print("1-Dimensional Array Shape" ,numbers.shape)

#2D Array
darray = np.array([
    [1,2,3,4],
    [5,6,7,8]]);
print("2-Dimensional Array",darray)
print(type(numbers))
print("2-Dimensional Array Shape" ,darray.shape)

```

```python
## Creating Special Arrays
zeros = np.zeros((3,4))
print(zeros)

ones = np.ones((2,5))
print(ones)

random_array = np.random.rand(3,3)
print("Random Values between 0 and 1")
print(random_array)

random_normal = np.random.randn(3,3)
print("Random Values from Normal Distribution")
print(random_normal)
```

```python
#### Indexing Numpy Arrays ####

x = np.array([10,20,30,40,50])
print("Array:" , x)
print("Length of Array:" , len(x))

print("First Element", x[0])
print("Last Element", x[-1])
print("First 3 Elements", x[:3])
print("Last 3 Elements", x[2:])
```
