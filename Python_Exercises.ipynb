```python
### Tuples: Tuples are used to store multiple items in a single variable ####
### Tuples are like lists but they are immutable ####

fruits = ("apple", "banana", "cherry")
print(fruits)
print(fruits[0])
print(fruits[1])
print(fruits[2])

# Mixed data types
user_profile = ("Alice", 30, True, 98.6)
print(user_profile)

# Empty tuple
empty_seq = ()
print(empty_seq)

# CRITICAL: Single-element tuple requires a trailing comma
single_item = ("orange",)  # Correct
print(single_item)

not_a_tuple = ("orange")   # Incorrect: This is just a string!
print(not_a_tuple)

# Tuple packing (parentheses are optional)
dimensions = 1920, 1080
print(dimensions)


### Accessing and Slicing Items
colors = ("red", "green", "blue", "yellow", "purple")

# Positive index
print(colors[0])   # Output: red

# Negative index
print(colors[-1])  # Output: purple

# Slicing (start:stop:step)
print(colors[1:4]) # Output: ('green', 'blue', 'yellow')


#### Built-in Tuple Methods

scores = (10, 20, 30, 20, 40, 20)

# 1. count(): Counts occurrences of a value
print(scores.count(20))  # Output: 3

# 2. index(): Finds the first index position of a value
print(scores.index(30))  # Output: 2


### Returning Multiple Values from Functions

def get_min_max(numbers):
    return min(numbers), max(numbers)
result = get_min_max([5, 2, 9, 1]) # Returns (1, 9)

print(get_min_max([5, 2, 9, 1]))

### Dictionaries

flight_prices = {("New York", "London"): 450,("Paris", "Tokyo"): 850}
print(flight_prices)
```

```python
#### Functions ##########
#### Functions allow us to reuse the code ##########

def add_numbers(a, b):
  result = a + b
  return result

result = add_numbers(4,5)
print(result)


#### Function to compute mean #####
def compute_mean(values) :
  total = 0
  for value in values:
    total = total + value
    mean = total/len(values)
    return mean


marks = [90,80,70,50,60]
marks_mean = compute_mean(marks)
print(marks_mean)
```

```python
### Classes
### Important for pytorch

class Student:
   def __init__(self,name, marks) :
      self.name= name
      self.marks = marks

   def display(self) :
     print("Name:" , self.name , "Marks:" , self.marks)


student1 = Student("Subbarao", 75)
student1.display()

```

```python
class SimpleModel:

   def __init__(self, weight,bias) :
     self.weight = weight
     self.bias = bias

   def forward(self,x) :
      y = self.weight *x + self.bias
      return y


model = SimpleModel(weight=10,bias=2)
x = 5
output = model.forward(5)
print("Input:" , x)
print("Output:" , output)
```
